# Univariate Analysis

```python
#Histograms for all features, for checking data distribution
wine_df.hist(figsize=(12, 10), bins=30)
plt.tight_layout()
plt.show()
```

```python
#Boxplots for detecting outliers
plt.figure(figsize=(12, 6))
sns.boxplot(data=wine_df)
plt.xticks(rotation=45)
plt.title("Boxplot of Features")
plt.show()
```

```python
#Countplot for checking data wine quality data distribution
sns.countplot(x='quality', data=wine_df)
plt.title('Distribution of Wine Quality')
plt.show()
```

# Bivariate Analysis

```python
# Correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(wine_df.corr(), annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.scatterplot(x='alcohol', y='quality', data=wine_df)
plt.title('Alcohol vs Quality')
plt.show()
```

```python
#Boxplot for cheking distribution of alcohol across different quality levels.
sns.boxplot(data=wine_df, x='quality', y='alcohol',)
plt.title('Alcohol Content by Wine Quality')
plt.show()
```

```python
#Facet Grids
g = sns.FacetGrid(wine_df, col="quality", height=5, aspect=1.5, col_wrap=3)
g.map(sns.histplot, "alcohol", kde=True)
g.set_titles(col_template="Quality {col_name}")
plt.show()
```

```python
wine_df['alcohol_bin'] = pd.cut(wine_df['alcohol'], bins=3, labels=["Low", "Medium", "High"])
sns.countplot(x='alcohol_bin', hue='quality', data=wine_df)
plt.show()
```
